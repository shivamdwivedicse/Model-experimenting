```python
df = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot','floors', 'waterfront', 'view', 'condition','sqft_basement', 'yr_built', 'yr_renovated','statezip']]
```

```python
import datetime
#To find current year we use 
current_year = datetime.datetime.now().year
```

```python
# 1. yr_built ko House_Age mein badal do 
# Formula: Current Year - Year Built
df['house_age'] = current_year - df['yr_built']
```

```python
# 2. yr_renovated ko is_renovated (Binary) mein badalna
# Agar saal 0 se bada hai toh 1 (Yes), nahi toh 0 (No)
df['is_renovated'] = df['yr_renovated'].apply(lambda x: 1 if x > 0 else 0)

# 3. sqft_basement ko has_basement (Binary) mein badalna
# Agar basement area 0 se bada hai toh 1, nahi toh 0
df['has_basement'] = df['sqft_basement'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df
```

```python
#now remove the which is getted transformed 
cols_drop = ['yr_built', 'yr_renovated', 'sqft_basement']
df = df.drop(columns=cols_drop)
```

```python
df
```

```python
# now statezip has both alphabet and number lets fix thatabs

# 'WA 98133' se '98133' nikalne ke liye
df['zipcode'] = df['statezip'].apply(lambda x: x.split(' ')[1])

# Isse integer mein convert karein taaki model ise number samjhe
df['zipcode'] = df['zipcode'].astype(int)

# Purana column drop kardein
df = df.drop('statezip', axis=1)
df
```

```python
#lets do standard scaling :-
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# In columns ko scale karna zaroori hai
scale_columns = ['sqft_living', 'sqft_lot', 'house_age', 'zipcode' , 'bedrooms' , "floors"]

df[scale_columns] = scaler.fit_transform(df[scale_columns])
```

```python
# 1. Sabse pehle apna original cleaned dataframe (jisne transformations ho chuki hain) save rakhein
df_original = df.copy()

# --- TASK A: Logistic Regression (Classification) ---
df_clf = df_original.copy()
# Category banayein
df_clf['price_category'] = pd.qcut(df_clf['price'], q=3, labels=[0, 1, 2])
# Price hata dein kyunki yahan target category hai
df_clf.drop('price', axis=1, inplace=True)


# --- TASK B: Regression (KNN, Linear Regression) ---
df_reg = df_original.copy()
# Yahan price category ki zaroorat nahi hai, sirf asli price rahegi
# (Agar aapne df_original mein category nahi banayi thi, toh yahan kuch drop karne ki zaroorat nahi)
```
